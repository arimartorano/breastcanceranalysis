```python
# Building pink Pairplot


sns.pairplot(breastcancerdf[['radius_mean', 'texture_mean', 'area_mean', 'diagnosis']], hue='diagnosis',)
color='pink'
sns.pairplot(breastcancerdf, hue='diagnosis', color='pink')
plt.suptitle('Pairplot of Selected Features')
plt.show()
```

```python
numeric_columns = [column for column in breastcancerdf.columns if breastcancerdf[column].dtype == 'float64']
print(numeric_columns)
```

```python
for column in numeric_columns:
    plt.figure(figsize=(12,8))
    sns.kdeplot(data=breastcancerdf, x=column, hue='diagnosis', palette="RdPu", fill=True)
    plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = breastcancerdf.corr()
sns.heatmap(corr, annot=True, cmap="RdPu_r")
```

```python


size = len(breastcancerdf['texture_mean'])
area = np.pi * (15 * np.random.rand(size))**2

# Set the color to pink (in this case, a light shade)
colors = ['pink'] * size

plt.xlabel("texture mean")
plt.ylabel("radius mean") 

plt.scatter(breastcancerdf['texture_mean'], breastcancerdf['radius_mean'], s=area, c=colors, alpha=0.5)
plt.show()
```
